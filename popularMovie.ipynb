# BroadCast

```python
def loadMovieNames():
    movieNames={}
    with open("///home/admin/Frank-Kanes/ml-100k/u.item") as f:
        for line in f:
            fields=line.split('|')
            movieNames[int(fields[0])]=fields[1]
        return movieNames
```

```python
nameDict=sc.broadcast(loadMovieNames())
```

```python
input=sc.textFile("file:///home/admin/Frank-Kanes/ml-100k/u.data")
```

```python
def normalizeMovie(input):
    line=input.split()
    return(int(line[1]))
```

```python
movie=input.map(normalizeMovie)
```

```python
movieCount=movie.map(lambda x: (x,1)).reduceByKey(lambda x ,y: x + y)
```

```python
movieSorted=movieCount.map(lambda x : (x[1] , x[0])).sortByKey()
```

```python
movieSortedNames=movieSorted.map(lambda (x, y) : (nameDict.value[y] , x))
```

```python
result=movieSortedNames.collect()
```

```python
for movie,count in result:
    print(str(movie) + " " + str(count))
```
